```python
from IPython.display import clear_output

def display_board(board):
    
    clear_output()
    
    print('     |     |     ')
    print('  '+board[1]+'  '+'|'+'  '+board[2]+'  '+'|'+'  '+board[3]+'  ')
    print('     |     |     ')
    print('-----------------')
    print('     |     |     ')
    print('  '+board[4]+'  '+'|'+'  '+board[5]+'  '+'|'+'  '+board[6]+'  ')
    print('     |     |     ')
    print('-----------------')
    print('     |     |     ')
    print('  '+board[7]+'  '+'|'+'  '+board[8]+'  '+'|'+'  '+board[9]+'  ')
    print('     |     |     ')

def player_input():
    
    marker = ''
    
    while marker != 'X' and marker != 'O':
        marker = input('Player 1, choose X or O: ')
    
    
    player1 = marker
    
    if player1 == 'X':
        player2 = 'O'
    else:
        player2 = 'X'
        
    
    return (player1,player2)

def place_marker(board,current_player,next_position):
     
    board[next_position] = current_player

def win_check(board, mark):
    
    if board[1] == board[2] == board[3] != ' ':
        return True
    elif board[4] == board[5] == board[6] != ' ':
        return True
    elif board[7] == board[8] == board[9] != ' ':
        return True
    elif board[1] == board[4] == board[7] != ' ':
        return True
    elif board[2] == board[5] == board[8] != ' ':
        return True
    elif board[3] == board[6] == board[9] != ' ':
        return True
    elif board[1] == board[5] == board[9] != ' ':
        return True
    elif board[7] == board[5] == board[3] != ' ':
        return True
    return False

import random

def choose_first():
    
    global current_player
    global current_playernum
    
    random_num = random.randint(1,2)
    
    if random_num % 2 == 0:
        current_player = player2_marker
        current_playernum = '2'
        print('Player 2 goes first!')
    else:
        current_player = player1_marker
        current_playernum = '1'
        print('Player 1 goes first!')

def space_check(board,next_position):
    
    if board[next_position] == ' ':
        return True
    return False

def full_board_check(board):
    
    for i in range(1,len(board)):
        if board[i] == ' ':
            return False
    return True

def player_choice(board):
    
    isitfree = False
    
    while isitfree == False:
        next_position = ' '
        while next_position != range(1,10):
            try:
                next_position = int(input("Please enter your next move (1-9): "))
                if space_check(board,next_position):
                    isitfree = True
                    break
                else:
                    print("This not a free space, please try again!")
            except:
                print("Please input an ACTUAL number (1-9)")
    return next_position

def replay():
    
    playagain = str(input("Would you like to play again? (Yas or Nah): "))

    while playagain != "Yas" and playagain != "Nah":
        playagain = str(input("Wrong input y'all (Yas or Nah): "))
    if playagain == "Yas":
        return True
    return False

print('Welcome to Tic Tac Toe!')

while True:

    player1_marker,player2_marker = player_input()
    
    board = [' ']*10
    current_player = ''
    current_playernum = ''
    next_position = 0
    
    display_board(board)
    
    choose_first()

    while True:
        print(f"It's Player {current_playernum}'s turn ({current_player})")
        
        next_position = player_choice(board)
        
        place_marker(board,current_player,next_position)
    
        display_board(board)
        
        if win_check(board,current_player):
            print(f'{current_player} won!')
            break
        
        if full_board_check(board):
            print("Womp womp. It's a tie!")
            break
        
        if current_player == 'X':
            current_player = 'O'
        elif current_player == 'O':
            current_player = 'X'
        if current_playernum == '1':
            current_playernum = '2'
        elif current_playernum == '2':
            current_playernum = '1'
            
    if not replay():
        break
```
